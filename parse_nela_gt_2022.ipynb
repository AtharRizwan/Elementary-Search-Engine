```python
import pandas as pd
import os
```

```python
os.listdir(r"newsdata")
```

```python
file_list = [os.path.join("newsdata", file) for file in os.listdir("newsdata")]
file_list
```

```python
dfs = []
for file in file_list:
    dfs.append(pd.read_json(file))
```

```python
df = pd.concat(dfs, ignore_index = True)
```

```python
del dfs
```

```python
df.head()
```

```python
df.drop(columns=["date", "published"], inplace = True)
```

```python
df.drop(columns=["published_utc", "collection_utc"], inplace = True)
```

```python
sampled_df = df.sample(n=200000)
```

```python
sampled_df.reset_index(inplace=True)
```

```python
sampled_df.drop(columns=["index"], inplace = True)
```

```python
sampled_df
```

```python
sampled_df.to_json("sampled_articles.json")
```
